```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [10,5]
img1 = cv2.imread('img1.jpg', cv2.IMREAD_COLOR)
plt.imsave('img1bgr.jpg', img1)
img1rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('img2.jpg', cv2.IMREAD_COLOR)
plt.imsave('img2bgr.jpg', img2)
img2rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3 = cv2.imread('img3.jpg', cv2.IMREAD_COLOR)
plt.imsave('img3bgr.jpg', img3)
img3rgb = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)

plt.subplot(131)
plt.imshow(img1rgb)
plt.axis('off')

plt.subplot(132)
plt.imshow(img2rgb)
plt.axis('off')

plt.subplot(133)
plt.imshow(img3rgb)
plt.axis('off')

plt.imsave('img1bgr.jpg',img1)
plt.imsave('img2bgr.jpg',img2)
plt.imsave('img3bgr.jpg',img3)

```

```python
plt.subplot(131)
plt.imshow(img1[:,:,2], cmap='gray')
plt.imsave('img1gray.jpg', img1[:,:,2], cmap='gray')
plt.axis('off')

plt.subplot(132)
plt.imshow(img2[:,:,2], cmap='gray')
plt.imsave('img2gray.jpg', img2[:,:,2], cmap='gray')
plt.axis('off')

plt.subplot(133)
plt.imshow(img3[:,:,2], cmap='gray')
plt.imsave('img3gray.jpg', img3[:,:,2], cmap='gray')
plt.axis('off')
```

```python
def showImgFormatColor(img, imgrgb):
  plt.rcParams['figure.figsize'] = [10,5]
  r = img[:,:,2]
  g = img[:,:,1]
  b = img[:,:,0]

  r_img = np.zeros_like(imgrgb)
  r_img[:,:,0] = r

  g_img = np.zeros_like(imgrgb)
  g_img[:,:,1] = g

  b_img = np.zeros_like(imgrgb)
  b_img[:,:,2] = b

  plt.subplot(141)
  plt.title('Imagen Original')
  plt.imshow(imgrgb)
  plt.axis('off')

  plt.subplot(142)
  plt.title('Canal Rojo')
  plt.imshow(r_img)
  plt.axis('off')

  plt.subplot(143)
  plt.title('Canal Verde')
  plt.imshow(g_img)
  plt.axis('off')

  plt.subplot(144)
  plt.title('Canal Azul')
  plt.imshow(b_img)
  plt.axis('off')

  plt.show()
  
```

```python
showImgFormatColor(img1, img1rgb)
```

```python
showImgFormatColor(img2, img2rgb)
```

```python
showImgFormatColor(img3, img3rgb)
```

```python
def showImgFalseColor(img):
  b,g,r = cv2.split(img)

  plt.subplot(141)
  plt.title('Imagen Original')
  plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
  plt.axis('off')
  
  plt.subplot(142)
  plt.title('Canal Rojo')
  plt.imshow(r, cmap='Reds')
  plt.axis('off')
  
  plt.subplot(143)
  plt.title('Canal Verde')
  plt.imshow(g, cmap='Greens')
  plt.axis('off')
  
  plt.subplot(144)
  plt.title('Canal Azul')
  plt.imshow(b, cmap='Blues')
  plt.axis('off')
  
```

```python
showImgFalseColor(img1)
```

```python
showImgFalseColor(img2)
```

```python
showImgFalseColor(img3)
```
