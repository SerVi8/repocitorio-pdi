```python
#Librerias
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

Ejercicio 1

```python
image1 = cv2.cvtColor(cv2.imread('./imagen1.jpg'), cv2.COLOR_BGR2RGB)
img_rec = image1[600:800, 600:850]

plt.rcParams['figure.figsize']=[10,5]

plt.subplot(1,2,1)
plt.imshow(image1)
plt.subplot(1,2,2)
plt.imshow(img_rec)

```

Ejercicio 2

```python
#definicion de colores
#colors = [(0,'black'),(1,'white')]
#colors = [(0,'#000000'), (0.4,[1,.3,0,1]), (1,'yellow')]
colors = [(0,'#440255'), (0.5,'#228b8d'), (1,'#fae622')]

#creacion de la imagen degradada
cmap = LinearSegmentedColormap.from_list('degradado', colors)
img_deg = cmap(np.linspace(0,0.9,255).reshape(-1,1))

plt.imshow(img_deg, aspect='auto')
```

Ejercicio 3

```python
plt.rcParams['figure.figsize']=[10,5]
img_size = 800
R = np.ones((img_size,img_size))*0
G = np.ones((img_size,img_size))*0
B = np.ones((img_size,img_size))*0

chees_board = np.dstack((R,G,B))
chees_board = np.array(chees_board, dtype=int)
white = [255, 255, 255]

for i in range(8):
  for j in range(8):
    if (i+j)%2 == 0:
      chees_board[i*100:i*100+100, j*100:j*100+100] = white


plt.figure()
plt.imshow(chees_board)
```

Ejercicio 5

```python
imgGray = cv2.cvtColor(image1, cv2.COLOR_RGB2GRAY)

colorsBW = [(0,'white'),(1,'black')]
cmap = LinearSegmentedColormap.from_list('degradado',colorsBW)

degra_l = cmap(np.linspace(0,1,100).reshape(-1,1))

img_degl = cv2.bitwise_and(degra_l,imgGray, mask=imgGray)


plt.rcParams['figure.figsize']=[10,5]
plt.subplot(1,2,1)
plt.imshow(imgGray, cmap='gray')
plt.title('Imagen original')

plt.subplot(1,2,2)
plt.imshow(img_degl)
```
