Ejercicio 1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

```python
plt.rcParams['figure.figsize']=[10,5]
img_size = 800
R = np.ones((img_size,img_size))*255
G = np.ones((img_size,img_size))*255
B = np.ones((img_size,img_size))*255

chees_board = np.dstack((R,G,B))
chees_board = np.array(chees_board, dtype=int)
black = [0,0,0]
gray = [128,128,128]

for i in range(8):
  for j in range(8):
    if (i+j)%2 != 0:
      chees_board[i*100:i*100+100, j*100:j*100+100] = gray
    if(i%2 ==0) and ((i+j)%2 == 0):
      chees_board[i*100:i*100+100, j*100:j*100+100] = black


plt.figure()
plt.imshow(chees_board)


```

```python
img = cv2.imread('./Imagen1.jpg', 0)
alto,ancho = img.shape[:2]

mascara = np.zeros_like(img)

centro = (ancho //2, alto//2)
radio = 150
cv2.circle(mascara, centro, radio, (255), -1)

img_mascara = cv2.bitwise_and(img, img, mask=mascara)

mascara2 = np.zeros_like(img)

img_mascara2 = cv2.bitwise_and(img, img, mask=mascara2)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(img_mascara2, cmap='gray')
```

```python
import random
img_rgb = cv2.cvtColor(cv2.imread('./Imagen1.jpg'),cv2.COLOR_BGR2RGB)
alto,ancho = img_rgb.shape[:2]

x=random.randint(0,alto)
y=random.randint(0,ancho)
img_rec = img_rgb[x:y, x:y]

img = cv2.imread('./Imagen1.jpg', 0)
win = 25
fact = 1
img_umbr = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, win, fact)
plt.figure(figsize=(10,5))

img_eca=cv2.normalize(img,None,0,255,cv2.NORM_MINMAX)
img_fil = cv2.applyColorMap(img_eca, cv2.COLORMAP_BONE)



plt.rcParams['figure.figsize']=[10,5]
plt.subplot(1,4,1)
plt.title('Imagen Original')
plt.imshow(img_rgb)
plt.imsave('./Imagenes/Ejercicio5-1.jpg', img_rgb)
plt.subplot(1,4,2)
plt.title('Imagen recorte')
plt.imshow(img_rec)
plt.imsave('./Imagenes/Ejercicio5-2.jpg', img_rec)
plt.subplot(1,4,3)
plt.title('Imagen Umbralizada')
plt.imshow(img_umbr, cmap='gray')
plt.imsave('./Imagenes/Ejercicio5-3.jpg', img_umbr)
plt.subplot(1,4,4)
plt.title('Imagen Con filtro')
plt.imshow(img_fil, cmap='gray')
plt.imsave('./Imagenes/Ejercicio5-4.jpg', img_fil)
```
