```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
img_rgb = cv2.cvtColor(cv2.imread('../Img/salamandra.png', cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB)

img_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)

img_rgb2 = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)

fig, axs = plt.subplots(1, 3, figsize=(20, 10))

axs[0].imshow(img_rgb)
axs[0].set_title('Imagen RGB')

axs[1].imshow(img_hsv)
axs[1].set_title('Imagen HSV')

axs[2].imshow(img_rgb2)
axs[2].set_title('Imagen RGB2')
```

```python
img_bgr = cv2.imread('../Img/llamitas.jpeg', 0)
_,umbralBinario = cv2.threshold(img_bgr, 127, 255, cv2.THRESH_BINARY)
_,umbralBinarioInverso = cv2.threshold(img_bgr, 127, 255, cv2.THRESH_BINARY_INV)

plt.subplot(1,3,1)
plt.imshow(img_bgr, cmap='gray')
plt.title('Imagen Original BGR')

plt.subplot(1,3,2)
plt.imshow(umbralBinario, cmap='gray')
plt.title('Imagen Binaria')

plt.subplot(1,3, 3)
plt.imshow(umbralBinarioInverso, cmap='gray')
plt.title('Imagen Binaria Inversa')

```

```python
img_yub = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2YUV)
img_rgb2 = cv2.cvtColor(img_yub, cv2.COLOR_YUV2RGB)

fig, axs = plt.subplots(1, 3, figsize=(20, 10))

axs[0].imshow(img_rgb)
axs[0].set_title('Imagen RGB')

axs[1].imshow(img_yub)
axs[1].set_title('Imagen YUB')

axs[2].imshow(img_rgb2)
axs[2].set_title('Imagen RGB2')
```
