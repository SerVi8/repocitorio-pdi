Filtros Bayer

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

img = cv2.cvtColor(cv2.imread('../Img/salamandra.png', cv2.IMREAD_COLOR), cv2.COLOR_BGR2GRAY)

img_bayer1 = cv2.cvtColor(img, cv2.COLOR_BayerBGGR2BGR)
img_bayer2 = cv2.cvtColor(img, cv2.COLOR_BayerGB2BGR)

plt.subplot(1,2,1)
plt.imshow(img_bayer1)

plt.subplot(1,2,2)
plt.imshow(img_bayer2)
```
