```python
#Importacion de Librerias

import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
plt.rcParams['figure.figsize'] = [10,5]
size_img = 100

R = np.ones((size_img, size_img)) + 62
G = np.ones((size_img, size_img)) + 120
B = np.ones((size_img, size_img)) + 226

lienzo = np.dstack((R, G, B))
lienzo = np.array(lienzo, dtype=int)



lienzo[0:50, 0:50] = (250, 0, 0)
lienzo[0:50, 50:100] = (0, 255, 0)
lienzo[50:100, 0:50] = (0, 0, 255)
lienzo[50:100, 50:100] = (0, 0, 0)
plt.imsave('img4cuadrantes.png', lienzo)


plt.figure()
plt.imshow(lienzo)
plt.show()
```

```python
img = cv2.imread('../Img/llamitas.jpeg', cv2.IMREAD_GRAYSCALE)

img_i =img.dtype
print('Profundidad del color:', img_i)
img_16bit= np.uint16(img)
new_img= img_16bit.dtype
print('Profundidad del color nueva:', new_img)

plt.subplot(121)
plt.imshow(img)

plt.subplot(122)
plt.imshow(img_16bit)
```

```python
fig, axs = plt.subplots(1,4, figsize=(25, 10))

img_brg = cv2.imread('../Img/salamandra.png')
img_rgb = cv2.cvtColor(img_brg, cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)

axs[0].imshow(img_rgb)
axs[0].set_title('imagen')

fig_img1 = axs[1].imshow(hsv[..., 0], cmap='gray', vmin=0, vmax=179)
fig.colorbar(fig_img1, ax=axs[1], shrink=0.3)
axs[1].set_title('Canal H')

fig_img2 = axs[2].imshow(hsv[..., 1], cmap='gray', vmin=0, vmax=179)
fig.colorbar(fig_img2, ax=axs[2], shrink=0.3)
axs[2].set_title('Canal S')

fig_img1 = axs[3].imshow(hsv[..., 2], cmap='gray', vmin=0, vmax=179)
fig.colorbar(fig_img1, ax=axs[3], shrink=0.3)
axs[3].set_title('Canal V')

```
