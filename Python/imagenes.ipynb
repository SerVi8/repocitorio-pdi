Canales y colores reales

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [10,5]
img = cv2.imread('mascotas.jpg', cv2.IMREAD_COLOR)
imgrgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

r = img[:,:,2]
g = img[:,:,1]
b = img[:,:,0]

r_img = np.zeros_like(imgrgb)
r_img[:,:,0] = r

g_img = np.zeros_like(imgrgb)
g_img[:,:,1] = g

b_img = np.zeros_like(imgrgb)
b_img[:,:,2] = b

plt.subplot(141)
plt.title('Imagen Original')
plt.imshow(imgrgb)
plt.axis('off')

plt.subplot(142)
plt.title('Canal Rojo')
plt.imshow(r_img)
plt.axis('off')

plt.subplot(143)
plt.title('Canal Verde')
plt.imshow(g_img)
plt.axis('off')

plt.subplot(144)
plt.title('Canal Azul')
plt.imshow(b_img)
plt.axis('off')


plt.show()
```

Colores falsos

```python
import cv2
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10,5]
img = cv2.imread('mascotas.jpg', cv2.IMREAD_COLOR)

b,g,r = cv2.split(img)

plt.subplot(141)
plt.title('Imagen Original')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(142)
plt.title('Canal Rojo')
plt.imshow(r, cmap='Reds')
plt.axis('off')

plt.subplot(143)
plt.title('Canal Verde')
plt.imshow(g, cmap='Greens')
plt.axis('off')

plt.subplot(144)
plt.title('Canal Azul')
plt.imshow(b, cmap='Blues')
plt.axis('off')


```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('llamitas.jpeg', cv2.IMREAD_COLOR)
imgrgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

print(f"Alto: {imgrgb.shape[0]}")
print(f"Ancho: {imgrgb.shape[1]}")
print(f"Canales: {imgrgb.shape[2]}")
fig, axs = plt.subplots(1,4, figsize = (20, 5))

axs[0].imshow(imgrgb)
axs[0].set_title('Imagen Original')

fig_img = axs[1].imshow(imgrgb[...,0 ], cmap='gray')
fig.colorbar(fig_img, ax=axs[1], shrink =0.5)
axs[1].set_title('Canal Rojo')

fig_img = axs[2].imshow(imgrgb[...,1 ], cmap='gray')
fig.colorbar(fig_img, ax=axs[2], shrink =0.5)
axs[2].set_title('Canal Verde')

fig_img = axs[3].imshow(imgrgb[...,2 ], cmap='gray')
fig.colorbar(fig_img, ax=axs[3], shrink =0.5)
axs[3].set_title('Canal Azul')

```
